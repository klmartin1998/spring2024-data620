```python
#Import necessary libraries

import networkx as net
import matplotlib.pyplot as plt
```

```python
g = net.Graph()
```

```python
nodes = ['Andre', 'Beverly','Carol', 'Diane','Ed','Fernando','Garth','Heather','Ike','Jane']
```

```python
g.add_nodes_from(nodes)
```

```python
edges = [('Andre','Beverly'),('Andre','Carol'),('Andre','Diane'),('Andre','Fernando'),('Beverly','Diane'),
         ('Beverly','Ed'),('Beverly','Garth'),('Carol','Diane'),('Carol','Fernando'),
         ('Diane','Ed'),('Diane','Fernando'),('Diane','Garth'),('Ed','Garth'),('Fernando','Garth'),
         ('Fernando','Heather'),('Garth','Heather'),('Heather','Ike'),('Ike','Jane')]
```

```python
g.add_edges_from(edges)
```

```python
g.edges()
```

```python
net.draw(g,with_labels=True, node_size=600)
```
